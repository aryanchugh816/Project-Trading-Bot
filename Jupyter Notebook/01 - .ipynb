### We will form a function which will take in parameters ==> (name_of_stock, time_interval, output_format) as input and save all the data in right format with a descriptive named CSV file


```python
def stock_data_to_csv(stock, interval="1min", output="compact", API_KEY="OMORWG3DQWN5BWUV"):
    from alpha_vantage.timeseries import TimeSeries
    
    ts = TimeSeries(key=API_KEY, output_format="pandas")
    
    data, meta_data = ts.get_intraday(symbol=stock, interval=interval, outputsize=output)
    
    data = data.iloc[::-1]
    
    name = stock+"_"+interval+"_"+output+".csv"
    data.to_csv(name)
    
    return
```

### We need to only plot certain columns/quantities: {"open", "close", "high", "low"} and not date or volume hence we will form a function that will take the name of file as input and what columns we want to plot and will plot a simple line graph for us

```python
def plot_line_graph(file="Data/Timeseries_alpha_vantage/AAPL_1min_compact.csv", cols=[2], separate=False):
    
    # We will give a Docstring indicating which column number belongs to what quantity:
    """cols: {1 : open, 2 : high, 3 : low, 4 : close, 5 : volume}"""
    
    data = pd.read_csv(file)
    
    columns = data.columns
    
    # Date column is of no use for us and for plotting purposes hence we will remove it from our data
    data = data.get(columns[1:])
    
    plt.style.use("seaborn")
    
    if not separate:
        plt.figure(figsize=(16,12))
        for i in cols:
            plt.plot(data[columns[i]], label=columns[i].split()[-1])

        plt.legend()
        plt.show()
        
    else:
        for i in cols:
            plt.figure(figsize=(16,12))
            plt.plot(data[columns[i]], label=columns[i].split()[-1])
            plt.legend()
            plt.show()
    
    return
    
```

### A simple Stacked LSTM model to predict stock prices

```python
from keras.models import Sequential
from keras.layers import *
from sklearn.preprocessing import MinMaxScaler
```

```python
# We will be using a stacked LSTM model for predicting prices

model = Sequential()
model.add(LSTM(4,input_shape=(3,1), return_sequences=True))
model.add(LSTM(4))
model.add(Dense(1))
model.compile(loss="mean_squared_error", optimizer='adam', metrics=['acc'])
model.summary()
```

```python
# LSTM takes input of the form: (samples, time-steps, features) so we will create a function that will take our X
# time series values and create X,Y values of the appropriate form

# We will have to input the dataset of shape: (x,1)

def create_dataset(dataset, look_back=1):
    
    X = []
    Y = []
    
    for i in range(len(dataset)-look_back):
        temp = dataset[i:(i+look_back),0]
        X.append(temp)
        Y.append(dataset[i+look_back,0])
    
    return np.array(X),np.array(Y)
```

```python
print(AAPL_1_full["1. open"].values.shape)
```

```python
# LSTM network gives the best outputs when we input normalized values/ values that belong to normal gaussian distribution

minmax = MinMaxScaler()
```

```python
data = AAPL_1_full["1. open"].values
```

```python
data = data.reshape(-1,1)
```

```python
data_m = minmax.fit_transform(data)
```

```python
data_m.shape
```

```python
X,Y = create_dataset(data_m, 3)
print(X.shape, Y.shape)
```

```python
print(X[:5], Y[:5])
```

```python
XT = X[:int(0.8*X.shape[0])]
YT = Y[:int(0.8*Y.shape[0])]

Xt = X[int(0.8*X.shape[0]):]
Yt = Y[int(0.8*Y.shape[0]):]

XT = XT.reshape(-1,3,1)
Xt = Xt.reshape(-1,3,1)

print(XT.shape, YT.shape)
print(Xt.shape, Yt.shape)
```

```python
hist = model.fit(XT, YT, epochs=50, batch_size=1, validation_split=0.2)
```

```python
y_pred = model.predict(Xt)
```

```python
plt.figure(figsize=(16,8))
plt.plot(Yt, label="Actual Data")
plt.plot(y_pred, label="Predicted data")
plt.legend()
plt.show()
```

```python
# Model accuracy:

print(model.metrics_names)
print(model.evaluate(Xt, Yt))

# These type of models show very less accuracy because evaluate function matches exact values 
```
